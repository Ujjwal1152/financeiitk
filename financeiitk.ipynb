```python
from urllib.request import urlretrieve
amzn_price_url = 'https://query1.finance.yahoo.com/v7/finance/download/AMZN?period1=1648515037&period2=1680051037&interval=1d&events=history&includeAdjustedClose=true'
urlretrieve(amzn_price_url,'AMZN.csv')
data = pd.read_csv('AMZN.csv')
data
```

```python
data.Date = pd.to_datetime(data.Date)
data.info()
```

```python
data = data.set_index('Date')
```

```python
mpf.plot(data)
```

```python
mpf.plot(data,type='line',volume=True)
```

```python
mpf.plot(data.loc['2022-04'],volume = True)
```

```python
mpf.plot(data.loc['2020-04':'2022-07'],type = "candle",mav=(20),volume = True)
```

The above graph depicts the moving average of amazon share prices for time period of 20 days.
